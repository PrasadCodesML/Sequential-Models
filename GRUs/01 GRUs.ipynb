### GRUs

GRU is an RNN architecture used for sequential inputs

Why GRU exist since the problem with long term context is solved by LSTM

The problem with LSTMs is the architecture is complex and contains more paramters due to which training is slow, but GRUs have a simpler architecture and have less parameters

In GRUs there are only 2 gates and LSTMs have 3 gates but still the performance is similar to LSTM


### Big Idea behind GRU
image.png
GRU have only one state and in this one hidden state it carries both long term context and short term context

Gates of GRUs - Reset gate, Update gate

Input to GRUs
1. Hidden state of previous input (t-1)
2. Current timestamp input

  - h(t-1) -> previous hidden state
  - h(t) -> Current hidden state
  - r(t) -> reset gate
  - z(t) -> update gate
  - h(t)' -> Condidate hidden state

All  the above have same dimension

- X(t) -> Current timestamp input

All the yellow boxes are neural network layers, and no of nodes is a hyperparameter

Steps to calculate h(t)
1. Calculate r(t) (reset gate)
2. Calculate h(t)' (candidate hidden state)
3. Calculate z(t) (update gate)
4. Calculate h(t) (current hidden state)





### Hidden state

Hidden state is nothing but memory and you store context in this memory (the vector stores some kind of context)

### Reset Gate

- Reset gate converts previous hidden state h(t-1) into candidate hidden state h(t)''
- The reset gate determines how much of the past information should be forgotten. It helps the GRU decide whether to reset the memory of the previous states for the current computation. A lower value of the reset gate means less past information is considered in the current update.


### Update Gate

- Update gate converts candidate hidden state into current hidden state h(t)
- The update gate controls how much of the previous memory is retained and how much of the new information is incorporated into the current state. It essentially balances between keeping the past and adding new information, similar to the forget and input gates in LSTMs combined.
